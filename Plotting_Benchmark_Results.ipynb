```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xvec = np.arange(100,800,100)
xvec2 = np.arange(100,2100,100)

plot_array_gauss = np.load('plot_array_gauss.npy')
plot_array_gauss2 = np.load('plot_array_gauss2.npy')

plot_array_ssp = np.load('plot_array_ssp.npy')
plot_array_ssp2 = np.load('plot_array_ssp2.npy')
```

```python
plt.xlabel('number of training points')
plt.ylabel('run time in seconds')
#plt.title('run times for different SMO algorithms, Gaussian')
sklearn, = plt.plot(xvec2,np.concatenate((plot_array_gauss[0],plot_array_gauss2[0])),label='scikit-learn')
wss1, = plt.plot(xvec2,np.concatenate((plot_array_gauss[1],plot_array_gauss2[1])),label='WSS1 with caching')
wss1noK, = plt.plot(xvec,plot_array_gauss[2],label='WSS1 without caching')
wss2, = plt.plot(xvec,plot_array_gauss[3],label='WSS2 with caching')
platt, = plt.plot(xvec,plot_array_gauss[4],label='Platt')
plt.legend(handles=[platt,wss1noK,wss2,wss1,sklearn])

plt.savefig('benchplot_gauss.pdf')
#plt.ylabel('some numbers')
plt.show()
```

```python
plt.xlabel('log(number of training points)')
plt.ylabel('log(run time in seconds)')
#plt.title('loglog run times for different SMO algorithms, Gaussian')

sklearn, = plt.loglog(xvec2,np.concatenate((plot_array_gauss[0],plot_array_gauss2[0])),label='scikit-learn')
wss1, = plt.loglog(xvec2,np.concatenate((plot_array_gauss[1],plot_array_gauss2[1])),label='WSS1 with caching')
wss1noK, = plt.loglog(xvec,plot_array_gauss[2],label='WSS1 without caching')
wss2, = plt.loglog(xvec,plot_array_gauss[3],label='WSS2 with caching')
platt, = plt.loglog(xvec,plot_array_gauss[4],label='Platt')
plt.legend(handles=[platt,wss1noK,wss2,wss1,sklearn])

plt.savefig('benchplot_gauss_loglog.pdf')
#plt.ylabel('some numbers')
plt.show()
```

```python
fit_sklearn = np.polyfit(np.log(xvec),np.log(plot_array_gauss[0]),1)
fit_wss1 = np.polyfit(np.log(xvec),np.log(plot_array_gauss[1]),1)
print(fit_wss1)
print(fit_sklearn)
print('polynomial order =',fit_wss1[0])
ratio = (np.exp(fit_wss1[1])) / (np.exp(fit_sklearn[1]))
print('ratio of run times wss1/sklearn =',ratio)
```

Clearly, all the SMO algorithms scale quadratically with the number of training points, but with very different coefficients $\alpha$ in $run time = \alpha n^2$. WSS1 with caching runs approx. 5.8 times longer than scikit-learn.
